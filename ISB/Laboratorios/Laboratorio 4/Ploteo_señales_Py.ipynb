```python
import matplotlib.pyplot as plt
import numpy as np



#-------PLOTEO DE LA SEAL EN REPOSO
array_reposo = np.genfromtxt("Reposo_EKG.txt",delimiter="	",skip_header = 2) #leemos el archivo txt

M = array_reposo[:,-2].shape[0] #longitud de la seal
valores=array_reposo[:,-2]

cantidad=np.size(valores);
#Siendo la frecuencia de muestreo del BITalino de 1000 Hz

tiempo=np.arange(0,cantidad)
tiempo=tiempo/1000

valores=(((valores/1024)-0.5)*3.3)/(1100)
valores=valores*1000
plt.figure(figsize=[12,4])
plt.plot(tiempo, valores,label="Seal en reposo")
plt.ylabel('ECG Signal (mV)')
plt.xlabel('Time (s)');
plt.title('Seal ECG en Reposo')
plt.xlim([5,8])
plt.grid(True)
```
