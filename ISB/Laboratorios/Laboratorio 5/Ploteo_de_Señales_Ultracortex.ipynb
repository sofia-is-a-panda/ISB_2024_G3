#Data obtenida de OpenBCI GUI

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import fftpack
import math

#-------PLOTEO DE LA SEAL EN REPOSO
array_reposo = np.genfromtxt("/content/OpenBCI-RAW-2024-04-26_09-55-06.txt",delimiter=",",skip_header=5) #leemos el archivo txt
Fs=250;

#Cargamos valores de los 8 canales en variables

scalefactor=0.02235
valores_canal0=array_reposo[:,1]
valores_canal0=valores_canal0*scalefactor

valores_canal1=array_reposo[:,2]
valores_canal1=valores_canal1*scalefactor

valores_canal5=array_reposo[:,6]
valores_canal5=valores_canal5*scalefactor

valores_canal6=array_reposo[:,7]
valores_canal6=valores_canal6*scalefactor

valores_canal7=array_reposo[:,8]
valores_canal7=valores_canal7*scalefactor

number_samples=valores_canal0.shape[0]
time=np.arange(0,number_samples)
time=time/Fs
plt.figure(figsize=[12,4])
plt.plot(time,valores_canal0)
plt.plot(time,valores_canal1)

plt.plot(time,valores_canal5)
plt.plot(time,valores_canal6)
plt.plot(time,valores_canal7)
plt.grid(True)
plt.ylabel("Microvoltios (uv)")

#plt.ylim([80000,88000])
plt.xlabel('Tiempo (s)')
plt.title("EEG Seales (Todo los canales)")
plt.legend(["Canal 0","Canal 1","Canal 5","Canal 6","Canal 7"])
plt.ylim([1250,2000])
plt.show()



```

###FFT de los distintos canales

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fftpack
```

### Canal 0

```python
# Parmetros de la seal
N = len(valores_canal0)
Ts = 1 / 250

# FFT
Y = fftpack.fft(valores_canal0)
Y_shifted = fftpack.fftshift(Y)  # Aplicar fftshift
angle = np.angle(Y_shifted)

# Frecuencias
Fs = 250
f = fftpack.fftfreq(Y.size, Ts)
f_shifted = fftpack.fftshift(f)  # Aplicar fftshift a las frecuencias

plt.plot(f_shifted, Y_shifted)
plt.xlim([-50, 50])
plt.ylim([0, 20000])
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.title('FFT del Canal 0')
plt.show()
```
